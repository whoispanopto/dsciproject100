### 5. Predictive Modeling: K-Nearest Neighbours Classification

The goal of modeling was to predict whether a player subscribed to the newsletter based on their gameplay and demographic characteristics. Following the workflow in *Data Science: A First Introduction*, we used a K-nearest neighbours (KNN) classifier implemented with **tidymodels** and the **kknn** engine.

##### 5.1 Data Preparation for Modeling

We first selected the predictors identified during EDA (`played_hours`, `Age`, and `experience`), removed rows with missing values, and encoded the outcome `subscribe` as a factor for classification.


```r
players_knn <- players |>
  select(subscribe, played_hours, Age, experience) |>
  drop_na(subscribe, played_hours, Age, experience) |>
  mutate(subscribe = as_factor(subscribe))
```

We then created an 80/20 train–test split, stratified by subscription status to preserve class proportions.


```r
set.seed(123)

players_split <- initial_split(players_knn, prop = 0.8, strata = subscribe)
players_train <- training(players_split)
players_test  <- testing(players_split)
```

##### 5.2 Preprocessing with a Recipe

Because KNN relies on distance calculations, numeric predictors must be on a comparable scale. We used a recipe to (i) create dummy variables for the categorical experience predictor and (ii) centre and scale all numeric predictors.


```r
players_recipe <- recipe(subscribe ~ played_hours + Age + experience,
                         data = players_train) |>
  step_dummy(experience) |>
  step_normalize(all_numeric_predictors())
```

##### 5.3 Model Specification

We selected k = 5 as a simple and commonly used neighbourhood size in introductory classification models, consistent with course materials. Because the goal was interpretability rather than optimizing predictive accuracy, we did not perform hyperparameter tuning.


```r
knn_spec <- nearest_neighbor(
  neighbors   = 5,
  weight_func = "rectangular"
) |>
  set_engine("kknn") |>
  set_mode("classification")
```

##### 5.4 Workflow Construction and Model Fitting

We combined preprocessing and modeling into a single workflow and fit the model on the training data.



```r
knn_workflow <- workflow() |>
  add_recipe(players_recipe) |>
  add_model(knn_spec)

knn_fit <- knn_workflow |>
  fit(data = players_train)
```

##### 5.5 Generating Predictions on the Test Set

We generated predicted class labels for the held-out test set. Evaluation metrics (accuracy and confusion matrix) are reported in the Results section.


```r
players_test_predictions <- predict(knn_fit, new_data = players_test) |>
  bind_cols(players_test)

players_test_predictions
```

## Results

Our exploratory analysis revealed several patterns that help explain which player characteristics are most predictive of subscribing to the game-related newsletter. Overall, subscription was common among players in the dataset, with 144 of 196 individuals subscribed. The distribution of gameplay activity and experience showed differences between subscribers and non-subscribers, allowing us to identify the features most strongly associated with subscription behaviour.

The most predictive characteristic that emerged from the exploratory findings was total played hours. Subscribers consistently accumulated substantially more gameplay time compared to non-subscribers. As shown in the bar plots, the mean played hours for subscribed players was dramatically higher, while the majority of non-subscribed players logged very little time on the server. The log-scaled histogram of total played hours further highlighted this disparity: although most players played only a few hours, the subset of high-activity players who formed the long right tail of the distribution were mostly subscribers. This indicates that long-term or highly involved players are far more likely to register for the newsletter.
A second predictor was experience level. Subscribers were more prevalent in every experience category. However, the discrepancies in subscription metrics by experience level are mostly described by the uneven proportions of player types in the data set. Nevertheless, most subscribed players are in the Amateur, Veteran and Regular player categories.
Demographic variables such as age showed minimal differences between the two groups. The average age varied only slightly between subscribers and non-subscribers, and no clear pattern suggested that age meaningfully influences subscription decisions. These findings indicate that player demographics are comparatively weak predictors of subscription behaviour relative to gameplay engagement metrics.

## Discussion

This analysis aimed to identify which of the player characteristics best set apart those who subscribed to a game-related newsletter. Among the descriptive statistics and the K-Nearest Neighbors classification model, Total Played Hours was the strongest predictor of subscription status. The subscribers of the newsletter were roughly 73% of the dataset, showing higher overall playtime than non-subscribers. This evidence shows that newsletter engagement is strongly linked with higher overall playtime in the Minecraft server.

These results are generally in line with expectations. It is intuitive to consider that the more hours players spend on the server, the more likely they are to want more information, updates, and community announcements via a newsletter. Generally speaking, high engagement increases the perceived value of supplementary content, which makes played hours a logical and reliable predictor. One finding that was a bit less expected was the lack of predictive power in Age. Although it would be reasonable to expect some differences in how players of different ages seek out information, especially in the context of online gaming communities, our findings show no meaningful variation. This suggests that newsletter engagement is influenced more by how players use the server than by their demographic characteristics.


The implications of these findings are meaningful to both the research team and possible server-level decision-making.
- Targeted marketing efforts can be optimized by centering communication on the high-engagement players who are statistically more likely to subscribe.
- Strategic communication can be adapted to motivate mid-engagement players to further develop their level of engagement.
- Audience profiling becomes more efficient if behavioral metrics, rather than demographic variables, guide outreach strategies. 

These results show that there are several useful next steps to explore in future research. A useful next step could be to check whether there is a specific amount of played hours after which players become much more likely to subscribe. Future research could also focus on identifying which in-game behaviors most strongly predict newsletter subscription. Another possible research direction would be to explore whether different types of players or playstyles show varying tendencies to subscribe, providing a deeper understanding of how different groups engage with the server.
